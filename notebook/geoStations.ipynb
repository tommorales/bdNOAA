```python
%matplotlib inline
import sys
sys.path.append("/Volumes/ekman/code/002_dev/bdNOAA") 
from src.app.controller import Controller
```

### Seleccionamos todas las estaciones disponibles

```python
def plot_all_stations():
    o = Controller()
    o.geo_station(all=True)


plot_all_stations()
```

### Seleccionamos por id de estación.

```python
def plot_id_stations(idStations):
    o = Controller()
    o.geo_station(idStations=idStations)
    
idStations = ["1611400", "1612340", "1615680"]
plot_id_stations(idStations)
```

### Seleccionamos estaciones por estado.

```python
def plot_state_station(stateStations):
    o = Controller()
    o.geo_station(stateStations=stateStations)
    
stateStations = ["HI", "NY", "ME"]    
plot_state_station(stateStations)
```
